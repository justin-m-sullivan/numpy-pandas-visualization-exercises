### 1. Use matplotlib to plot the following equation:

- y = x^2 − x + 2
- You'll need to write the code that generates the x and y points.
- Add an anotation for the point 0, 0, the origin.  

```python
%matplotlib inline
```

```python
plt.figure(figsize = (10,8))

x = range(1,10)
y = [n ** 2 - n + 2 for n in x]

plt.plot(x,y, color = 'red', marker = 'x')

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^2 - x + 2$', fontsize = 14)


plt.title('$y = x^2 − x + 2$', fontsize = 20) 

plt.annotate('Origin', xy=(0,0), xytext=(-.15,.10), arrowprops={'facecolor': 'blue'})
plt.tight_layout()
plt.show()
```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense): 
- y=√x
- y = x^3
- y = 2^x
- y = 1 / (x + 1)
#You can use functions from the math module to help implement some of the equations above.

```python
plt.figure(figsize = (10,6))

x = range(1,10)
     
y4 = [math.sqrt(n) for n in x]
     
plt.plot(x,y4)
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$√X)$', fontsize=16)
plt.show()
```

```python
x = range(1,10)
y3 = [n ** 3 for n in x]
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$x^3$', fontsize=16)
plt.scatter(x,y3)
plt.title('$y = x ^ 3$', fontsize=20)
plt.show()
```

```python
x = range(1,10)
y2 = [2 ** n for n in x]

plt.plot(x,y2)
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$2 ^ x$', fontsize=16)
plt.title('$y = 2 ^ x$', fontsize=20)
plt.show()
```

```python
x = range(1,10)
y1 = [1 / (n + 1) for n in x]

plt.plot(x,y1)
plt.ylabel('$1 / (x + 1)$', fontsize=16)
plt.xlabel('$x$', fontsize=16)
plt.title('$y = 1 / (n + 1)$', fontsize=20)
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
ys = [y1, y2, y3, y4]
plt.suptitle('$y = y(x)$', fontsize=20)

for i in range(1,5):
    plt.subplot(2,2,i)
    plt.plot(ys[i-1])
    plt.tight_layout()
    plt.xlabel('x')
    plt.ylabel('$y = fx(x)$')
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (10,8))

kwargs = dict(ls = '--')
x = range(1,10)

plt.plot(x,y1, **kwargs, label=('$1 / (x + 1)$'))
plt.plot(x,y2, **kwargs, label=('$2 ^ x$'))
plt.plot(x,y3, **kwargs, label=('$x^3$'))
plt.plot(x,y4, **kwargs, label=('$√X$'))

plt.xlabel('$x$', fontsize=16)
plt.ylabel('$y(x)$', fontsize=16)
plt.title('$y = y(x)$', fontsize=20)
plt.legend()
plt.show()
```
