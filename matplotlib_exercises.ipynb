### 1. Use matplotlib to plot the following equation:

- y = x^2 âˆ’ x + 2
- You'll need to write the code that generates the x and y points.
- Add an anotation for the point 0, 0, the origin.  

```python
%matplotlib inline
```

```python
plt.figure(figsize = (8,4))

x = range(1,10)
y = [n ** 2 - n + 2 for n in x]

plt.plot(x,y, color = 'red', marker = 'x')

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^2 - x + 2$', fontsize = 14)

plt.show()
# plt.title('y = x^2 âˆ’ x + 2', fontsize = 20) -- Why doesnt title work?

plt.annotate("This is the Origin", xy = (0,0), xytext = (0.03,0.03), arrowprops = {})

```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense): 
- y=âˆšx

#You can use functions from the math module to help implement some of the equations above.

```python
plt.figure(figsize = (10,6))

x = range(1,10)
     
y4 = [math.sqrt(n) for n in x]
     
plt.plot(x,y4)
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$SQRT(X)$', fontsize=16)
plt.show()
```

```python
x = range(1,10)
y3 = [n ** 3 for n in x]
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$x^3$', fontsize=16)
plt.scatter(x,y3)
plt.show()
```

```python
x = range(1,10)
y2 = [2 ** n for n in x]

plt.plot(x,y2)
plt.show()
```

```python
x = range(1,10)
y1 = [1 / (n + 1) for n in x]

plt.plot(x,y1)
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (10,8))

kwargs = dict(ls = '--')
x = range(1,10)

plt.plot(x,y1, **kwargs)
plt.plot(x,y2, **kwargs)
plt.plot(x,y3, **kwargs)
plt.plot(x,y4, **kwargs)

plt.xlabel('$x$', fontsize=16)
plt.ylabel('$y(x)$', fontsize=16)

plt.show()
```
