### Curves to graph:

- a. y = 0x+1 and label the curve "O(1)"
- b. y = log(x) and label the curve "O(log n)"
- c. y = x and label the curve "O(n)"
- d. y = x âˆ— log(x) and label it "O(n log n)"
- e. y = x^2 and label it "O(n^2)"
- f. y = 2^x and label it "O(2^n)"
- g. y = n! and label it "O(n!)"
- h. y = n^n and label it "O(n^n)"

```python
#### Notebook set-up:
import matplotlib.pyplot as plt
import math
import numpy as np
from random import randint
import random
import pandas as pd
```

```python
plt.figure(figsize= (10,8))

#define x values:
x = range(1,3)

# define curves as ys
ya = [0*n+1 for n in x]
yb = [math.log(n) for n in x]
yc = [n for n in x]
yd = [n * math.log(n) for n in x]
ye = [n ** 2 for n in x]
yf = [2 ** n for n in x]
yg = [math.factorial(n) for n in x]
yh = [n ** n for n in x]



plt.plot(x,ya, label='$O(1)$')
plt.plot(x,yb, label='$O(log n)$')
plt.plot(x,yc, label='$O(n)$')
plt.plot(x,yd, label='$O(n log n)$')
plt.plot(x,ye, label='$O(n^2)$')
plt.plot(x,yf, label='$O(2^n)$')
plt.plot(x,yg, label='$O(n!)$')
plt.plot(x,yh, label='$O(n^n)$')
plt.title('Big O Notation', fontsize=20)
plt.xlabel('elements', fontsize=16)
plt.ylabel('operations', fontsize=16)
plt.legend()
plt.text(1.9, 0 ,'Justin Sullivan')

plt.show()


```
